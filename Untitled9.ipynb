```python
def TextAnalysis(text):
  def CharCount(mytext,char):
   return str(mytext).count(char)
  def wordlen(word):
    return len(word)
  def wordClean(word):
    for char in "^%$#@&**!?,." :
      word=word.replace(char," ")
    return len (word)
  specialcharDic={}
  wordDic={}
  wordReplaceDic={}
  for i in "*&^%$##@.,?":
    specialcharDic[i]=CharCount(text,i)
  for word in text.split() :
    wordDic[word]=wordlen(word)
    wordReplaceDic[word]=wordClean(word)

  return specialcharDic,wordDic,wordReplaceDic
```

```python
TextAnalysis(phrase)
```
